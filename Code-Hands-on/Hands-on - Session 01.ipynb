```python
# We create a star network

g2 = nx.Graph()
g2.add_node(100)
for i in range(50):
    g2.add_node(i)
    g2.add_edge(100,i)
    
    
```

```python
# returns a list with all nodes
g2.nodes()
```

```python
# Returns a dictionary with the degree (value) of each node (key)
g2_degrees = g2.degree()
```

```python
g2_degrees
```

```python
# crete a list with all the degrees
degrees = np.array([ g2_degrees[i] for i in g2.nodes() ])
```

# Load a network

```python
# Extension of the dataset is gml
g_f = nx.read_gml("data/graph_facebook.gml")
g_fdegrees = g_f.degree()
```

```python
# PLot the degree distribution
```

```python
degrees = [ g_fdegrees[i] for i in g_f.nodes() ]
```

```python
plt.hist( degrees )
plt.xlabel('degree')
plt.ylabel('count')
```

```python
_ = plt.hist(degrees, bins = 100, density = True)
```

```python
_ = plt.hist(degrees, bins = [0.5,2,10,20,50,100,200,400,800,1600], density = True)
```

```python
np.logspace(np.log10(1),np.log10(1024),11)
```

```python
my_bins = np.logspace(np.log10(min(degrees)),np.log10(max(degrees) +1), 30 ) 
_ = plt.hist(degrees, bins = my_bins,density = True)
plt.xscale('log')
plt.yscale('log')
```
