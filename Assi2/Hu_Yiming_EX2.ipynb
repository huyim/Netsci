```python
#1. Plot the ave. degree
def ave_deg(name, name_r, graph, graph_r):
    #ave. nearest neighbor degree
    k = nx.average_degree_connectivity(graph)
    k_r = nx.average_degree_connectivity(graph_r)

    #Plot two graphs together
    figure, axis = plt.subplots(1, 2, figsize=(14,5))
    
    axis[0].scatter(k.keys(), k.values())
    axis[0].set_title(name)
    
    #??Why not shown??
    plt.xlabel('Degree k')
    plt.ylabel('k_nn(k)')  

    axis[1].scatter(k_r.keys(), k_r.values())
    axis[1].set_title(name_r)

    plt.xlabel('Degree k')
    plt.ylabel('k_nn(k)')    
```

```python
#A01.2 Clustering and randomisation
#1. Plot the degree distribution in double-logscale and mark with a vertical line the average degree 

def deg_distribution(name, graph):
    #Load degrees
    g_degrees = graph.degree()
    degrees = [ g_degrees[i] for i in graph.nodes() ]

    #Compute ave. degres
    # sum = 0
    # for i in degrees:
    #     sum += degrees[i]
    # ave_degree = sum / graph.number_of_nodes()
    # A simple way:
    ave_degree = 2 * graph.number_of_edges() / graph.number_of_nodes()

    #Draw the graph
    my_bins = np.logspace(np.log10(min(degrees)),np.log10(max(degrees) +1), 30 )
    _ = plt.hist(degrees, bins = my_bins, density = True)
    plt.title(name)
    plt.xscale('log')
    plt.yscale('log')
    plt.axvline(ave_degree, label='Average degrees', c='m')

    #Format vertical line
    # colors = ['g', 'c', 'm']
    # for p, c in zip(xpoints, colors):
    #     plt.axvline(p,  label='line: {}'.format(p), c=c)
```

```python
#2. For each node, compute the clustering coefficient in the graphs  (denoted by ()) and  (denoted as ())
#3. Do a scatter plot of () vs. ()

def cluster(name, graph, graph_r):
    c = nx.clustering(graph)
    c_r = nx.clustering(graph_r)

    cc = []
    cc_r = []
    for i in c:
        cc.append(c[i])
    for i in c_r:
        cc_r.append(c_r[i])
    
    plt.scatter(cc, cc_r)
    plt.title(name)

    plt.xlabel('C(i)')
    plt.ylabel('Crnd(i)')  
```
