```python
#1. Plot the ave. degree
def ave_deg(name, graph):
    k = nx.average_degree_connectivity(graph)
    
    plt.scatter(k.keys(), k.values())
    plt.title(name)
    plt.xlabel('Degree k')
    plt.ylabel('k_nn(k)')    
```

```python
print("Blue: Real; Orange: random")
ave_deg("AstroPh: ", g_astroph)
ave_deg("Real & Random_AstroPh: ", grnd_astroph)
```

```python
#Function for Q2, Q3
def print_asso(name, g):
    print(name + str(nx.degree_assortativity_coefficient(g)))      
```

```python
#2. Compute the assortativity coefficient of the real network
print_asso("AstroPh: ", g_astroph)
print_asso("Celegans: ", g_celegans)
print_asso("Chess: ", g_chess)
print_asso("CondMat: ", g_cond)
print_asso("Dolphins: ", g_dolphins)
print_asso("Eu_airlines: ", g_eu)
print_asso("Facebook: ", g_facebook)
print_asso("Florentine_familes: ", g_florentine)
print_asso("Game_thrones: ", g_game)
print_asso("Internet: ", g_internet)
print_asso("Jazz_collab: ", g_jazz)
print_asso("Terrorists: ", g_terrorists)
```

```python
#3. Compute the assortativity coefficient of the randomised network
print_asso("Random_AstroPh: ", grnd_astroph)
print_asso("Random_Celegans: ", grnd_celegans)
print_asso("Random_Chess: ", grnd_chess)
print_asso("Random_CondMat: ", grnd_cond)
print_asso("Random_Dolphins: ", grnd_dolphins)
print_asso("Random_Eu_airlines: ", grnd_eu)
print_asso("Random_Facebook: ", grnd_facebook)
print_asso("Random_Florentine_familes: ", grnd_florentine)
print_asso("Random_Game_thrones: ", grnd_game)
print_asso("Random_Internet: ", grnd_internet)
print_asso("Random_Jazz_collab: ", grnd_jazz)
print_asso("Random_Terrorists: ", grnd_terrorists)
```
